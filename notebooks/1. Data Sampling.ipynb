# Data Sampling


```python
#Imports
import pandas as pd
import pyarrow.parquet as pq
import pyarrow as pa
import glob
from datetime import date
```

```python
#Read csv file (change directory in your path or renamve csv to trips)
df_trips = pd.read_csv("../data/trips.csv")
```

```python
cols_dropped = df_trips.dropna(inplace=False)
```

```python
print("The original frame has",len(df_trips),"rows")
print(len(df_trips)-len(cols_dropped),"null values within the stream")
print("Remember: There are also null values in columns just because of privacy reasons - Actually we don't should drop them")
```

```python
# Sampling function
def sampleFile(sample_size,df):
    df_new = df.sample(frac = (sample_size/100))
    return df_new
```

```python
#Sample df to 20% of original frame 
df_sample = sampleFile(20,df_trips)
```

```python
#Write parquet of sample frame
frame = pa.Table.from_pandas(df_sample)
pq.write_table(frame, '../data/df_sample_{}.parquet'.format(date.today().strftime("%d_%m_%Y")))
```
